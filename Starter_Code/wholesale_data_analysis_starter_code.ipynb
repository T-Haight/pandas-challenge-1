## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df[['unit_price','qty','subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['shipping_price'] = df.apply(lambda x: x['unit_weight'] * x['qty'] * 7 if x['unit_weight'] * x['qty'] > 50 else x['unit_weight'] * x['qty'] * 10, axis=1)
df[['unit_price','unit_weight','qty','shipping_price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price'] = df['subtotal'] + df['shipping_price'] + (df['subtotal'] + df['shipping_price']) * 0.0925
df['line_price'] = df['line_price'].round(2)
df[['subtotal','shipping_price','line_price']].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['line_price'] - df['line_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
df.loc[df['client_id'] == 33615, 'qty'].sum()
```

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_5_clients:
    total_spent = df.loc[df['client_id'] == client, 'line_price'].sum()
    print(f"{client}: ${total_spent:.2f}")
```

```python
# What I thought the previous code block was asking for (got the top 5 clients by quantity, calculated the total spent by each, then sorted by their total spent in descending order)
clients = df['client_id'].unique()
top_clients_total = []
for client in clients:
    qty_total = df.loc[df['client_id'] == client, 'qty'].sum()
    total_spent = df.loc[df['client_id'] == client, 'line_price'].sum()
    top_clients_total.append({'client_id': client, 'qty': qty_total, 'spent': total_spent})
top_clients_total = pd.DataFrame(top_clients_total)
top_clients_total = top_clients_total.sort_values(by='qty', ascending=False)
top_clients_total.head()
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_clients_summary = {'client_id': [], 'total_units_purchased': [], 'total_shipping_price': [], 'total_revenue': [], 'total_profit': []}
for client in top_5_clients:
    total_units_purchased = df.loc[df['client_id'] == client, 'qty'].sum()
    total_shipping_price = df.loc[df['client_id'] == client, 'shipping_price'].sum()
    total_revenue = df.loc[df['client_id'] == client, 'line_price'].sum()
    total_profit = df.loc[df['client_id'] == client, 'profit'].sum()
    top_5_clients_summary['client_id'].append(client)
    top_5_clients_summary['total_units_purchased'].append(total_units_purchased)
    top_5_clients_summary['total_shipping_price'].append(total_shipping_price)
    top_5_clients_summary['total_revenue'].append(total_revenue)
    top_5_clients_summary['total_profit'].append(total_profit)
top_5_clients_summary = pd.DataFrame(top_5_clients_summary)
top_5_clients_summary
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_5_clients_summary.rename(columns={'client_id': 'Client ID', 'total_units_purchased': 'Units', 'total_shipping_price': 'Shipping', 'total_revenue': 'Total Revenue', 'total_profit': 'Total Profit'}, inplace=True)

# Define the money columns. 
money_columns = ['Shipping', 'Total Revenue', 'Total Profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
        return value/1000000

# Apply the currency_format_millions function to only the money columns. 
for column in money_columns:
    top_5_clients_summary[column] = top_5_clients_summary[column].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
top_5_clients_summary.rename(columns={'Shipping': 'Shipping (Millions)', 'Total Revenue': 'Total Revenue (Millions)', 'Total Profit': 'Total Profit (Millions)'}, inplace=True)
top_5_clients_summary

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
top_5_clients_summary_sorted = top_5_clients_summary.sort_values(by='Total Profit (Millions)', ascending=False)
top_5_clients_summary_sorted
```
